```python
table = pd.DataFrame({
    'Unique':myntra.nunique(),
    'Null':myntra.isna().sum(),
    'NullPercent':myntra.isna().sum() / len(myntra),
    'Type':myntra.dtypes.values
})
display(table)
```

```python
myntra['BrandName'] = myntra['BrandName'].apply(lambda t: t.lower())
myntra['BrandName'] = myntra['BrandName'].apply(lambda t: t.capitalize())
```

```python
myntra["Individual_category"] = myntra["Individual_category"].apply(lambda t: t.lower())
myntra["Individual_category"] = myntra["Individual_category"].apply(lambda t: t.capitalize())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
features = ["DiscountPrice (in Rs)","OriginalPrice (in Rs)"]
for data in features:
    sns.boxplot(x=myntra[data])
    plt.tight_layout()
    plt.show()
```

```python
myntra.columns
```

```python
columns = ["Reviews","Ratings","OriginalPrice (in Rs)","DiscountPrice (in Rs)"]
for data in columns:
    plt.figure(figsize=(18,7))
    sns.histplot(x=myntra[data])
    plt.xticks(fontsize=14)
#     plt.tight_layout()
    plt.show()
```

```python
columns = ["Reviews","Ratings","DiscountPrice (in Rs)","OriginalPrice (in Rs)"]

for i in columns:
    plt.figure(figsize=(14,7))
    sns.kdeplot(x=myntra[i], hue=myntra["category_by_Gender"])
    plt.xticks(fontsize=14)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(14,7))
myntra['category_by_Gender'].value_counts().plot.pie(autopct='%.2f', textprops={'size':14})
```

```python
columns = ['BrandName','DiscountOffer', 'SizeOption', 'Ratings']

plt.figure(figsize=(10,20))
for i, col in enumerate(columns):
    axes = plt.subplot(4,2, i + 1)
    sns.countplot(y=myntra[col], order=myntra[col].value_counts().index[0:10], palette='Spectral')
    plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
